```python
fixed_effects= "log_net_sales_volume~log_price + month_April+ month_August+ month_December+ month_February+ month_January +  month_July+ month_June+ month_March+ month_May+month_November+ month_October+ month_September + Peak_on_New_Year+ Dip_on_New_Year + Dip_on_Thanksgiving + Peak_on_Thanksgiving+ Dip_1_week_after_Thanksgiving + Peak_1_week_after_Thanksgiving+ Dip_2_week_after_Thanksgiving+ Peak_2_week_after_Thanksgiving + Dip_on_Christmas + Peak_on_Christmas+ Dip_in_2nd_week_of_year + Peak_1_week_after_Halloween + Peak_on_4th_of_July + covid_period+ covid_flag_3 + covid_flag_4"
random_effects="~log_acv_wtd_distribution + log_price + acv_display_by_100+ acv_feature_and_display_by_100+acv_tpr_by_100 + acv_feature_by_100 +trend_term+pl1+pl2+pl3"
```

```python
from sklearn.metrics import mean_absolute_percentage_error, r2_score
```

```python
def calculate_mape(y_true, y_pred):
    return mean_absolute_percentage_error(y_true, y_pred)
def calculate_r2(y_true, y_pred):
    return r2_score(y_true, y_pred)
```

```python
grouped = df.groupby('ppg_retailer')
```

```python
mape_scores = grouped.apply(lambda x: calculate_mape(x['log_net_sales_volume'], x['predictions']))
r2_scores = grouped.apply(lambda x: calculate_r2(x['log_net_sales_volume'], x['predictions']))
```

```python
results_df = pd.DataFrame({
    'MAPE': mape_scores,
    'R2 Score': r2_scores
})

print(results_df)
```

```python
results_df = pd.DataFrame({
    'MAPE': mape_scores,
    'R2 Score': r2_scores
})
```

```python
mape_threshold = 0.3  
r2_threshold = 0.45  
```

```python
results_df['MAPE'] = results_df['MAPE'] <= mape_threshold
results_df['R2 Score'] = results_df['R2 Score'] >= r2_threshold
```

```python
print(results_df)
```

```python
coefficients_data = {'ppg_retailer': [], 'price_coefficient': []}

for retailer_id in df['ppg_retailer'].unique():
    retailer_df = df[df['ppg_retailer'] == retailer_id]
    price_coefficient = mdf.params['log_price']
    coefficients_data['ppg_retailer'].append(retailer_id)
    coefficients_data['price_coefficient'].append(price_coefficient)
coefficients_df = pd.DataFrame(coefficients_data)
```

```python
result_df = results_df.merge(coefficients_df, on='ppg_retailer')
```

```python
def is_coefficient_between_range(coeff):
    return -5 <= coeff <= -0.5
result_df['Beta'] = result_df['price_coefficient'].apply(lambda x: is_coefficient_between_range(x))
```
